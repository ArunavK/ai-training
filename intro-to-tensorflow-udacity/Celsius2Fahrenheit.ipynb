# Predicting the conversion algorithm from $^oF$ to $^oC$ using a simple neural network

## Importing required modules

```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
```

```python
from jupyterthemes import jtplot
jtplot.style('onedork', figsize=(16, 9))
```

## Generating the labelled input data

The Celsius value is the feature. The Fahrenheit value is the label.

```python
celsius_feature = np.array([-40, -10, 0, 8, 15, 22, 38], dtype=float)
fahrenheit_label = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)
```

## Creating layers for the neural network

```python
l0 = tf.keras.layers.Dense(units=1, input_shape=[1])
```

## Generating the model

```python
model = tf.keras.Sequential([l0])

model.compile(loss='mean_squared_error', optimizer=tf.keras.optimizers.Adam(0.1))
```

## Training the model

```python
history = model.fit(celsius_feature, fahrenheit_label, epochs=500, verbose=False)
print("Finished training the model")
```

```python
plt.xlabel('Epoch number')
plt.ylabel('Loss')
plt.plot(history.history['loss'])
plt.show()
```

```python
print("These are the learnt weights: {}".format(l0.get_weights()))
```
